```python
%sql
SELECT MONTH(salary_date) AS month_no, 
       date_format(salary_date, 'MMMM') AS report_date, 
       round(SUM(total_salary_value),2) AS total_salary_value 
FROM salary_silver 
GROUP BY MONTH(salary_date), date_format(salary_date, 'MMMM') 
ORDER BY MONTH(salary_date);
```

```python
%sql
-- Step 0: Create Schema if not exists
CREATE SCHEMA IF NOT EXISTS salary_data_project.gold_db;

-- drop table salary_data_project.gold_db.salary_gold;
-- Step 1: Create Gold Table if not exists
CREATE TABLE IF NOT EXISTS salary_data_project.gold_db.salary_gold (
  department_name VARCHAR(255),
  total_salary_value FLOAT,
  avg_salary FLOAT,
  report_date DATE
);

-- Step 2: Insert Aggregated Business Metrics
INSERT INTO salary_data_project.gold_db.salary_gold (department_name, total_salary_value, avg_salary, report_date)
SELECT  
  es.department_name,
  round(SUM(es.total_salary_value),2) AS total_salary_value,
  round(AVG(es.salary_amount),2) AS avg_salary,
  CURDATE() AS report_date
FROM salary_silver es
JOIN department_silver ds 
  ON es.department_name = ds.department_name
GROUP BY es.department_name;

SELECT * from gold_db.salary_gold;
```
